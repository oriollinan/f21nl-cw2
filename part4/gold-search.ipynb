# Gold Search

```python
import sys
from pathlib import Path
from dataclasses import dataclass

import pandas as pd
import plotly.express as px
```

```python
# Set correct working directory for imports
sys.path.append('../part2')
from part2.predict import predict
```

```python
CHECKPOINT=Path("nmt.model")
CONFIG=Path("..", "part2", "model_config.yaml")
VOCAB=Path("..", "part2", "vocab", "vocab.json")
ENCODER=Path("..", "part2", "vocab", "spm.model")
DECODING_STEP=70
SENTENCE="Two boys are sitting on a couch that is red and black."
VISUALISE=False
SAMPLE=True
```

```python
@dataclass
class Args:
    checkpoint_path: Path
    model_config: Path
    vocab_file: Path
    encoder: Path
    max_decoding_time_step: int
    sentence: str
    visualise: bool
    do_sample: bool
    top_k: int | None
    top_p: float | None
    temperature: float
```

```python
DEFAULT_ARGS = Args(
    checkpoint_path=CHECKPOINT,
    model_config=CONFIG,
    vocab_file=VOCAB,
    encoder=ENCODER,
    max_decoding_time_step=DECODING_STEP,
    visualise=VISUALISE,
    sentence=SENTENCE,
    do_sample=SAMPLE,
    top_k=None,
    top_p=None,
    temperature=1.0
)
```

```python
predict(DEFAULT_ARGS)
```

```python
TOP_K_VALUES=[50, 100, 150, 200]
TOP_P_VALUES=[0.2, 0.4, 0.6, 0.8]
TEMPERATURE_VALUES=[0.25, 0.5, 0.75, 1.0]
```

```python
grid = []
for k in TOP_K_VALUES:
    for p in TOP_P_VALUES:
        for t in TEMPERATURE_VALUES:
            grid.append((k, p, t))
```

```python
GOLD="Deux garçons sont assis sur un canapé qui est rouge et noir.</s>"
```

```python
correct = []
for k, p, t in grid:
    args = DEFAULT_ARGS
    args.top_k = k
    args.top_p = p
    args.temperature = t
    prediction = predict(args)
    if prediction == GOLD:
      correct.append((k, p, t))
```
